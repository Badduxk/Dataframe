```python
import pandas as pd

# 1. Membuat DataFrame
print("1. Membuat DataFrame dari data jumlah produksi sampah berdasarkan Kabupaten/Kota di Jawa Barat.")

data = {
    "Kabupaten/Kota": [
        "Kabupaten Bogor", "Kabupaten Sukabumi", "Kabupaten Cianjur", "Kabupaten Bandung", "Kabupaten Garut",
        "Kabupaten Bogor", "Kabupaten Sukabumi", "Kabupaten Cianjur", "Kabupaten Bandung", "Kabupaten Garut",
    ],
    "Tahun": [
        2018, 2018, 2018, 2018, 2018,
        2019, 2019, 2019, 2019, 2019
    ],
    "Produksi Sampah (ton)": [
        1716.8, 415.5, 965.19, 2068.06, 473.89,
        2914.65, 1003.64, 1115.81, 1334.12, 1144.01
    ]
}

# Membuat DataFrame
df = pd.DataFrame(data)
print(df)

# 2. Total Produksi Sampah di Tahun Tertentu
tahun_tertentu = 2018
total_sampah_tahun = sum(row['Produksi Sampah (ton)'] for i, row in df.iterrows() if row['Tahun'] == tahun_tertentu)
print(f"\n2. Total produksi sampah di Jawa Barat pada tahun {tahun_tertentu}: {total_sampah_tahun} ton")

# 3. Total Produksi Sampah per Tahun
total_per_tahun = {}
for i, row in df.iterrows():
    total_per_tahun[row['Tahun']] = total_per_tahun.get(row['Tahun'], 0) + row['Produksi Sampah (ton)']
print("\n3. Total produksi sampah per tahun:")
for tahun, total in total_per_tahun.items():
    print(f"{tahun}: {total:.2f} ton")

# 4. Total Produksi Sampah per Kota/Kabupaten per Tahun
total_per_kota_tahun = {}
for index, row in df.iterrows():
    kota = row['Kabupaten/Kota']
    tahun = row['Tahun']
    produksi = row['Produksi Sampah (ton)']

    if kota not in total_per_kota_tahun:
        total_per_kota_tahun[kota] = {}

    if tahun in total_per_kota_tahun[kota]:
        total_per_kota_tahun[kota][tahun] += produksi
    else:
        total_per_kota_tahun[kota][tahun] = produksi

hasil_tabel_kota = []

for kota, tahun_data in total_per_kota_tahun.items():
    for tahun, total in tahun_data.items():
        hasil_tabel_kota.append({'Kabupaten/Kota': kota, 'Tahun': tahun, 'Total Produksi (ton)': total})

df_hasil_kota = pd.DataFrame(hasil_tabel_kota)
df_total_per_tahun = pd.DataFrame(total_per_kota_tahun)

print("\nJumlah produksi sampah per Kota/Kabupaten per tahun:")
print(df_hasil_kota)

# Ekspor ke CSV dan Excel
df.to_csv('produksi_sampah.csv', index=False)
df.to_excel('produksi_sampah.xlsx', index=False)
df_total_per_tahun.to_csv("total_per_tahun.csv", index=False)
df_total_per_tahun.to_excel("total_per_tahun.xlsx", index=False)
df_hasil_kota.to_csv("total_per_kota_per_tahun.csv", index=False)
df_hasil_kota.to_excel("total_per_kota_per_tahun.xlsx", index=False)
print("\nData berhasil diekspor ke CSV dan Excel!")
```
